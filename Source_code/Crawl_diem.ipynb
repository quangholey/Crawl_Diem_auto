```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
import pandas as pd
import time
from datetime import datetime  
import os
for i in kyhieumatruong:
    # Đường dẫn tới ChromeDriver
    service = Service(r"C:\Users\QUANGHONEY\Desktop\Code\NCKH_Crawl_Data\chromedriver-win64\chromedriver.exe")
    driver = webdriver.Chrome(service=service)

    # Mở trang web
    driver.get('https://diemthi.tuyensinh247.com')

    # Tìm ô input và nhập mã trường
    input_field = driver.find_element(By.CSS_SELECTOR, 'input[type="text"]')
    input_field.clear()
    input_field.send_keys(f'{i}')
    input_field.send_keys(Keys.RETURN)

    # Đợi để xem kết quả tìm kiếm
    time.sleep(1)

    try:
        time.sleep(1)  # Đợi thêm một lát để tiện drop quảng cáo thủ công

        # Tìm liên kết đầu tiên trong danh sách kết quả hoặc có href chứa từ khóa nào đó
        links = driver.find_elements(By.XPATH, "//ul[@id='benchmarking']//a")

        # Nhấp vào liên kết đầu tiên phù hợp
        if links:
            first_link = links[0]

            # Cuộn đến phần tử trước khi nhấp
            driver.execute_script("arguments[0].scrollIntoView();", first_link)
            time.sleep(1)  # Đợi thêm một lát để tiện drop quảng cáo thủ công

            # Chuyển đến tab khác để tránh kích hoạt quảng cáo
            driver.execute_script("window.open(arguments[0].href, '_blank');", first_link)
            driver.switch_to.window(driver.window_handles[-1])

            # Đợi để trang mới tải
            time.sleep(5)

            try:
                # Tìm tất cả các div chứa bảng và nhãn liên quan
                divs = driver.find_elements(By.XPATH, "//div[@class='tables_point']/div")

                # Tạo một danh sách để lưu dữ liệu
                data = []

                # Lấy năm hiện tại
                current_year = datetime.now().year

                # Duyệt qua từng div để lấy label và bảng tương ứng
                for div in divs:
                    # Lấy nhãn (label) trong div
                    label = div.find_element(By.XPATH, ".//label[@class='title_point']")
                    title_text = label.text
                    data.append([current_year, title_text])  # Thêm năm hiện tại và label vào dữ liệu

                    # Tìm tất cả các hàng trong bảng điểm chuẩn trong cùng div
                    rows = div.find_elements(By.XPATH, ".//table//tr")

                    # Duyệt qua từng hàng và lấy dữ liệu
                    for row in rows:
                        # Lấy tất cả các cột trong hàng
                        cols = row.find_elements(By.TAG_NAME, 'td')
                        # Nếu hàng có dữ liệu thì thêm vào danh sách
                        if len(cols) > 0:
                            # Lấy văn bản từ mỗi cột
                            row_data = [col.text for col in cols]
                            # Thêm năm hiện tại vào hàng
                            row_data.insert(0, current_year)
                            # Thêm dữ liệu vào danh sách
                            data.append(row_data)

                # Chuyển dữ liệu thành DataFrame pandas không có tiêu đề cột
                df = pd.DataFrame(data)

                # Đường dẫn thư mục lưu tệp
                folder_name = f'C:/Users/QUANGHONEY/Desktop/Code/NCKH_Crawl_Data/Data_Tuyen_sinh_247/diem_{i}_cac_nam_vnexpress'

                # Kiểm tra xem thư mục đã tồn tại chưa, nếu chưa thì tạo mới
                if not os.path.exists(folder_name):
                    os.makedirs(folder_name)

                # Lưu DataFrame vào tệp CSV
                df.to_csv(f'{folder_name}/diem_chuan_{i}_all.csv', index=False, header=False, encoding='utf-8-sig')
                print(f"Dữ liệu đã được lưu vào tệp diem_chuan_{i}_all.csv")

            except Exception as e:
                print(f"Đã xảy ra lỗi khi lấy dữ liệu: {e}")
        else:
            print("Không tìm thấy liên kết nào phù hợp.")

    except Exception as e:
        print(f"Đã xảy ra lỗi: {e}")
    finally:
        # Đóng trình duyệt
        driver.quit()
```
